- Program to accept three inetegers and print the largest of three. Make use of only if statement

- **Program that inputs three numbers and calculates two sums as per this:**
  - *Sum 1 as the sum of all input numbers*
  - *Sum2 as the sum of non-duplicate numbers; if there are duplicate numbers in the input, ignores them*

```python
# First Approach

sum1 = sum2 =0
n1 = int(input("Enter first number: "))
n2 = int(input("Enter second number: "))
n3 = int(input("Enter third number: "))
sum1 = n1 + n2 + n3
if n1 != n2 and n1 != n3:
    sum2 += n1
if n2 != n1 and n2 != n3:
    sum2 += n2
if n3 != n1 and n3 != n2:
    sum2 += n3
print("Numbers are", n1, n2, n3)
print("Sum of three given numbers is", sum1)
print("Sum of non-duplicate numbers is", sum2)
```

```python
# Second Approach

sum1 = sum2 = 0
num1 = int(input("Enter number 1: "))
num2 = int(input("Enter number 2: "))
num3 = int(input("Enter number 3: "))
sum1 = num1 + num2 + num3
if num1 == num2: 
    if num3 != num1: 
        sum2 += num3
else:
    if num1 == num3: 
        sum2 += num2
    else: 
        if num2 == num3: 
            sum2 += num1
        else:
            sum2 = num1 + num2 + num3
print("Numbers are:", num1, num2, num3)
print("Sum of three given numbers is:", sum1)
print("Sum of non-duplicate numbers is:", sum2)
```

- Program to find the multiples of a number out of given five numbers.

```python
print("Enter the five numbers below")
n1 = float(input("Enter number 1: "))
n2 = float(input("Enter number 2: "))
n3 = float(input("Enter number 3: "))
n4 = float(input("Enter number 4: "))
n5 = float(input("Enter number 5: "))
divisor = float(input("Enter the divisor: "))
count = 0
print("Multiple of", divisor, "are:")
remainder = n1 % divisor
if remainder == 0:
    print(n1,sep="")
    count += 1
remainder = n2 % divisor
if remainder == 0:
    print(n2,sep="")
    count += 1
remainder = n3 % divisor
if remainder == 0:
    print(n3,sep="")
    count += 1
remainder = n4 % divisor
if remainder == 0:
    print(n4,sep="")
    count += 1
remainder = n5 % divisor
if remainder == 0:
    print(n5,sep="")
    count += 1
print("Total multiples of", divisor, "are:", count)
```

- Right-angled triangle pattern

```python
for i in range(1, 21):
    for j in range(i):
        print("G ", end=" ")
    print()
```

- Inverted right-angled triangle

```python
for i in range(20, 0, -1):
    for j in range(i):
        print("G ", end=" ")
    print()
```

- Pyramid pattern

```python
n = 20
for i in range(1, n + 1):
    print(" " * (n - i), end="")
    print("G " * i)
```

- Inverted pyramid

```python
n = 20
for i in range(n, 0, -1):
    print(" " * (n - i), end="")
    print("G " * i)
```

- Square pattern

```python
for i in range(20):
    for j in range(50):
        print("G", end=" ")
    print()
```

- Diamond pattern

```python
n = 15

for i in range(1, n + 1):
    print(" " * (n - i) + "G " * i)

for i in range(n - 1, 0, -1):
    print(" " * (n - i) + "G " * i)
```

- Hollow square

```python
n = 25
for i in range(n):
    for j in range(n):
        if i == 0 or i == n-1 or j == 0 or j == n-1:
            print("G", end=" ")
        else:
            print(" ", end=" ")
    print()
```

- X Pattern

```python
n = 25
for i in range(n):
    for j in range(n):
        if i == j or i + j == n - 1:
            print("G", end=" ")
        else:
            print(" ", end=" ")
    print()
```

- Write Python code to add the odd numbers up to (and including) a given value N and print the result

- Write a program to input a number and then print its first and last digit raised to the length of the number (the number of digits in the number is the length of the number).

```python
num = input("Enter a number: ")
length = len(num)
first_digit = int(num[0])
last_digit = int(num[-1])
print("Length of number:", length)
print("First digit raised to length:", first_digit ** length)
print("Last digit raised to length:", last_digit ** length)
```
