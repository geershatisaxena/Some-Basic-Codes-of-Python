- **Given a dictionary M which stores the marks of the students of class with roll numbers as the keys and marks as the values. Write a program to check if anyone has scored marks as 89.9.**

```python
# Given dictionary of roll numbers and marks
M = {
    101: 78.5,
    102: 89.9,
    103: 67.0,
    104: 92.3,
    105: 81.4
}

# Check for marks 89.9
if 89.9 in M.values():
    print("Yes, a student has scored 89.9 marks.")
else:
    print("No student has scored 89.9 marks.")
```

- **Write a program to create a dictionary M which stores the marks of the students of class with roll numbers as the keys and marks as the values. Get the number of students as input.**

```python
M = {}   # empty dictionary
n = int(input("Enter number of students: "))
for i in range(n):
    roll = int(input("Enter roll number: "))
    marks = int(input("Enter marks: "))
    M[roll] = marks

print("Dictionary of Roll Numbers and Marks:")
print(M)
```

- **Consider dictionary M created in the previous example. Write a program to modify the marks of a roll number. Obtain the roll number as input**

```python
# Program to modify marks of a given roll number

roll = int(input("Enter roll number to modify marks: "))

if roll in M:
    new_marks = int(input("Enter new marks: "))
    M[roll] = new_marks
    print("Marks updated successfully.")
else:
    print("Roll number not found.")

print("Updated Dictionary:")
print(M)
```

- **Write a program to create a dictionary namely Numerals from following two lists.**

- **Write a program to add new students' roll numbers and marks in the dictionary M created with roll numbers as the keys and marks as the values.**

```python
# Existing dictionary
M = {
    1: 85,
    2: 78,
    3: 90
}

# Input number of new students
n = int(input("Enter number of new students: "))

# Adding new students
for i in range(n):
    roll = int(input("Enter roll number: "))
    marks = int(input("Enter marks: "))
    M[roll] = marks

# Display updated dictionary
print("Updated Dictionary M:")
print(M)
```
