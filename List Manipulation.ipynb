- Program to print elements of a list [ GEERSHATI SAXENA) in separate lines along with element's both indexes (positive and negative).

```python
lst = ["GEERSHATI", "SAXENA"]
length = len(lst)
for i in range(length):
    print("Element:", lst[i])
    print("Positive Index:", i)
    print("Negative Index:", i - length)
    print()
```
