- Extract two list-slices out of a given list of numbers. Display and print the sum of elements of first list-slice which contains every other element of the list between indexes 5 to 15. Program should also display the average of elements in second list slice that contains every fourth element of the list. The given list contains numbers from 1 to 20.

```python
# Given list from 1 to 20
lst = list(range(1, 21))

# First slice: every other element between index 5 and 15
slice1 = lst[5:16:2]
sum_slice1 = sum(slice1)

# Second slice: every fourth element of the list
slice2 = lst[::4]
avg_slice2 = sum(slice2) / len(slice2)

# Display results
print("Original List:", lst)
print("\nFirst list slice (every other element from index 5 to 15):", slice1)
print("Sum of first list slice:", sum_slice1)

print("\nSecond list slice (every fourth element):", slice2)
print("Average of second list slice:", avg_slice2)
```

- Write a program that asks the user to input a number a list to be appended to an existing list. Whether the user enters a single number or a list of numbers, the program should append the list accordingly

```python
# Existing list
lst = [1, 2, 3, 4, 5]

# Ask user for input
user_input = input("Enter a number or a list of numbers separated by commas: ")

# Check if user entered multiple numbers (comma-separated)
if ',' in user_input:
    # Convert input into a list of integers
    new_items = [int(x.strip()) for x in user_input.split(',')]
    lst.extend(new_items)  # Append all numbers
else:
    # Single number, convert to int and append
    lst.append(int(user_input))

# Display the updated list
print("Updated list:", lst)
```

- Write a program that displays options for deleting elements in a list with option for : if element is to be deleted with value or by using its position or a list slice is to be deleted.

```python
val = [17, 23, 18, 19]
print("The list is:", val)
print("\nDeletion Menu")
print("1. Delete using Value")
print("2. Delete using index")
print("3. Delete a sublist")

dch = int(input("\nEnter choice: "))

if dch == 1:
    item = int(input("Enter item to be deleted: "))
    try:
        val.remove(item)
        print("List now is:", val)
    except ValueError:
        print(f"Item {item} not found in the list.")

elif dch == 2:
    try:
        index = int(input("Index of item to be deleted: "))
        val.pop(index)
        print("List now is:", val)
    except IndexError:
        print(f"Invalid index! List has only {len(val)} elements.")
    except ValueError:
        print("Please enter a valid integer index.")

elif dch == 3:
    try:
        l = int(input("Lower limit of list slice: "))
        h = int(input("Upper limit of list slice: "))
        
        # Check if indices are valid
        if l < 0 or h > len(val) or l >= h:
            print(f"Invalid slice range! Must be 0 <= lower < upper <= {len(val)}")
        else:
            del val[l:h]
            print("List now is:", val)
    except ValueError:
        print("Please enter valid integer values.")
    except TypeError:
        print("Invalid input type!")

else:
    print("Valid choices are 1/2/3 only.")
```

- Write a program to input two lists and display the maximum element from the elements of both the lists combined, along with its index in its list.

```python
# Input two lists
list1 = list(map(int, input("Enter elements of first list separated by space: ").split()))
list2 = list(map(int, input("Enter elements of second list separated by space: ").split()))

# Combine both lists
combined = list1 + list2

# Find maximum element
max_element = max(combined)

# Find index and list where max element lies
if max_element in list1:
    index = list1.index(max_element)
    print("Maximum element:", max_element)
    print("Found in FIRST list at index:", index)
else:
    index = list2.index(max_element)
    print("Maximum element:", max_element)
    print("Found in SECOND list at index:", index)
```

- Write a program to find the largest/smallest value of asset in a given list of assets of a company.

```python
def analyze_assets(assets):
    # Edge case: Check if the list is empty
    if not assets:
        return None, None

    # Find the asset with the highest value
    # key=lambda x: x['value'] tells Python to compare based on the 'value' key
    most_expensive = max(assets, key=lambda x: x['value'])

    # Find the asset with the lowest value
    least_expensive = min(assets, key=lambda x: x['value'])

    return most_expensive, least_expensive

# --- Example Data ---
company_assets = [
    {'name': 'Office Building', 'value': 500000},
    {'name': 'High-end Server', 'value': 15000},
    {'name': 'Office Chair', 'value': 150},
    {'name': 'Company Car', 'value': 25000},
    {'name': 'Coffee Machine', 'value': 800}
]

# --- Run the Program ---
largest, smallest = analyze_assets(company_assets)

if largest and smallest:
    print(f"Largest Asset: {largest['name']} (${largest['value']})")
    print(f"Smallest Asset: {smallest['name']} (${smallest['value']})")
else:
    print("No assets found.")
```

- ABC company's assets, liabilities and capital details are available in the form of three lists. Write a program to test if the accounting equation holds true for ABC company or not.

```python
assets = [50000, 30000, 20000]
liabilities = [40000, 15000, 5000]
capital = [10000, 15000, 15000]
total_assets = sum(assets)
total_liabilities = sum(liabilities)
total_capital = sum(capital)
print("Total Assets:", total_assets)
print("Total Liabilities:", total_liabilities)
print("Total Capital:", total_capital)

if total_assets == total_liabilities + total_capital:
    print("✅ Accounting equation holds true for ABC company.")
else:
    print("❌ Accounting equation does NOT hold true for ABC company.")
```

- Given are the lists containing various sales and costs amounts for two companies ABC co. and XYZ co. Write a program to calculate the return-on-sales (ROS) for both the companies and display which company's management successfully cuts costs while increasing revenue. Formula to calculate ROS is as given below:

```python
# Sales and costs lists for both companies
abc_sales = [50000, 40000, 30000]
abc_costs = [30000, 25000, 20000]

xyz_sales = [60000, 45000, 35000]
xyz_costs = [42000, 32000, 28000]

# Calculate totals using list manipulation
total_abc_sales = sum(abc_sales)
total_abc_costs = sum(abc_costs)

total_xyz_sales = sum(xyz_sales)
total_xyz_costs = sum(xyz_costs)

# Calculate ROS
ros_abc = ((total_abc_sales - total_abc_costs) / total_abc_sales) * 100
ros_xyz = ((total_xyz_sales - total_xyz_costs) / total_xyz_sales) * 100

# Display ROS values
print("ABC Company ROS:", round(ros_abc, 2), "%")
print("XYZ Company ROS:", round(ros_xyz, 2), "%")

# Compare and display result
if ros_abc > ros_xyz:
    print("ABC company's management is more successful in cutting costs and increasing revenue.")
elif ros_xyz > ros_abc:
    print("XYZ company's management is more successful in cutting costs and increasing revenue.")
else:
    print("Both companies have equal management efficiency.")
```
